```python
# 📝 Task 8:
# From the `customers` table, count how many customers exist in each state.
# Show two columns:
#    - customer_state
#    - total_customers
# Sort the results in descending order of total_customers
query = """
SELECT customer_state, count(customer_id) as total_customers
FROM customers
GROUP BY customer_state
ORDER BY total_customers  DESC
LIMIT 10 ;
"""
df = pd.read_sql_query(query, conn)
df
```

```python
# 📝 Task 9:
# Find the top 10 most frequently ordered product categories.
# You’ll need to join:
#    - `order_items` (has product_id)
#    - `products` (has product_category_name)
#    - `product_category_name_translation` (has English names)
# Show:
#    - product_category_name_english
#    - total_orders
# Sort in descending order of total_orders

query = """
Select count(o.product_id) as total_orders,c.product_category_name_english
from order_items o
    join products p on o.product_id = p.product_id
    join product_category_name_translation c on p.product_category_name = c.product_category_name
    group by c.product_category_name_english
    order by total_orders desc
    limit 10
"""
df = pd.read_sql_query(query, conn)
df
```

```python
# Count how many orders exist for each order_status in the `orders` table.
# Show:
#   - order_status
#   - total_orders
# Sort by total_orders DESC
query = """
SELECT order_status, COUNT(order_id) AS total_orders
FROM orders
GROUP BY order_status
ORDER BY total_orders  DESC
LIMIT 10 ;
"""
df = pd.read_sql_query(query, conn)
df
```

```python
# From the `orders` table, count how many orders were delivered in the year 2018.
# Show:
#   - order_status
#   - total_orders
# Only consider orders where:
#   - order_delivered_customer_date is in 2018
# Sort in descending order of total_orders

query = """
SELECT
    c.customer_state,
    COUNT(o.order_id) AS total_order
FROM
    orders o
JOIN
    customers c ON o.customer_id = c.customer_id
WHERE
    SUBSTR(o.order_delivered_customer_date, 1, 4) = '2018'
GROUP BY
    c.customer_state
ORDER BY
    total_order DESC
LIMIT 10;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#Top 5 Product Categories Ordered in 2018
query = """
SELECT count(o.order_id) as total_order,p.product_category_name
FROM
    orders o
JOIN
    order_items i on o.order_id=i.order_id
join
    products p on i.product_id = p.product_id
    where substr(o.order_delivered_customer_date,1,4)="2018"
group by p.product_category_name
order by  total_order desc
LIMIT 5 ;
"""
#desc comes with order by not limit
df = pd.read_sql_query(query, conn)
df
```

```python
#Monthly Order Trend in 2018
query = """
SELECT substr(order_purchase_timestamp,1,7) as month ,count(order_id) as total_order,order_purchase_timestamp
FROM orders
    where substr(order_purchase_timestamp,1,4)="2018"
    group by month
order by month  ;
"""
df = pd.read_sql_query(query, conn)
df
```
